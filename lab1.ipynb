# Load the dataset

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('CDS Elevator Arrivals (Responses) - Form Responses 1.csv')
print(data) # rmv later

counts = [data['Elevator ID'].value_counts()[i] for i in range(1, 7)]
columns = [1, 2, 3, 4, 5, 6]
```

# Visualiztion 1

```python
plt.xlabel("Elevator")
plt.ylabel("Occurences")
plt.bar(columns,counts)
```

This helps the viewer compare the number of occurrances of each elevator.

# Visualiztion 2

```python
plt.pie(counts, labels=columns)
```

This Pie chart shows what portion of arrivals each elevator is. It looks like they all have about 1/6 of the arrivals, with some slight variation.

# Visualiztion 3

```python
plt.xlabel("Elevator")
plt.ylabel("Frequency")
plt.imshow((counts,), cmap='YlGn', interpolation='nearest')
```

This heatmap tells the viewer which elevators come the most (lighter colors), and which come the least.
